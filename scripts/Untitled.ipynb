```python
def load_data():
    dtype = np.float32

    pas_accum1 = pickle.load(open("pas_accum_1.pkl", "rb")).values
    pas_accum2 = pickle.load(open("pas_accum_2.pkl", "rb")).values
    pas_accum3 = pickle.load(open("pas_accum_3.pkl", "rb")).values

    pas_accum1 = pas_accum1.astype(dtype)
    pas_accum2 = pas_accum2.astype(dtype)
    pas_accum3 = pas_accum3.astype(dtype)

    x1, y1 = pas_accum1[:, :-1], pas_accum1[:, -1]
    x2, y2 = pas_accum2[:, :-1], pas_accum2[:, -1]
    x3, y3 = pas_accum3[:, :-1], pas_accum3[:, -1]

    X = np.vstack([x1, x2, x3])
    Y = np.hstack([y1, y2, y3])

    x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

    return x_train, x_test, y_train, y_test

x_train, x_test, y_train, y_test = load_data()
N, n_x = x_train.shape
# Standardize data
x_train, x_test, mean_x_train, std_x_train = dataset.standardize(x_train, x_test)
y_train, y_test, mean_y_train, std_y_train = dataset.standardize(y_train, y_test)
```
